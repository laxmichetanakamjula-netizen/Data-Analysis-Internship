```python
import pandas as pd
```

Step-1:Data Access & Familiarization

Understanding the Dataset take and seperating the column headers, datatypes, checking for the missing values, showing the unique values

```python
df=pd.read_csv("Internship/train.csv")
df.head()
print(df.columns)
print(df.dtypes) 
print(df.isnull().sum()) 
print(df.nunique())
```

Creating the data dictonary for the above dataset concentrating more about each variable's meaning, type, and potential business relevance.

```python
data_dictionary=pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes,
    "Missing Values": df.isnull().sum(),
    "Unique Values": df.nunique()
})
data_dictionary.to_csv("data_dictionary.csv", index=False)
print(data_dictionary)
```
